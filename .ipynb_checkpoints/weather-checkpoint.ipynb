```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import normalize
```

```python
df = pd.read_excel(r'sthlm_weather_1900-2020.xlsx') # From: http://miljobarometern.stockholm.se/klimat/klimat-och-vaderstatistik/manadsmedeltemperatur/
```

```python
df.head(3)
```

```python
df = df.rename(columns={"Measure area": "Month"})

df_piv = df.pivot(index="Month", columns="Year", values="Value").reindex(["Januari", 
                                                                            "Februari",
                                                                            "Mars",
                                                                            "April",
                                                                            "Maj", 
                                                                            "Juni",
                                                                            "Juli",
                                                                            "Augusti",
                                                                            "September",
                                                                            "Oktober",
                                                                            "November",
                                                                            "December"])

df_piv
```

```python
normed_df = normalize(df_piv, axis=1, norm='l1')
normed_df = pd.DataFrame(normed_df)
normed_df
```

```python
# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(50, 20))
sns.heatmap(df_piv, annot=True, linewidths=.5, ax=ax);
```

```python
sns.set(style="ticks")

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(50, 10))

# Plot the orbital period with horizontal boxes
sns.boxplot(x="Year", y="Value", data=df,
            whis=1.5, color="white", linewidth=2)

# Add in points to show each observation
sns.swarmplot(x="Year", y="Value", data=df,
              size=2, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

```python
df["Decade"]=df["Year"]//10*10
df.sample(3)
```

```python
sns.set(style="ticks")

# Initialize the figure
f, ax = plt.subplots(figsize=(20, 10))

# Plot the orbital period with horizontal boxes
sns.boxplot(x="Decade", y="Value", data=df,
            whis=1.5, color="white", linewidth=2)

# Add in points to show each observation
sns.swarmplot(x="Decade", y="Value", data=df,
              size=2, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
plt.title('Mean Temperatures per Monthly Observation and Decade');
plt.ylabel('Monthly Mean Temperatures at Observatorielunden');
```

```python
sns.set(style="ticks")

# Show the results of a linear regression within each dataset
sns.lmplot(x="Year", y="Value", col="Month", hue="Month", data=df,
           col_wrap=3, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1});
```

```python
# From https://climate.nasa.gov/vital-signs/global-temperature/

df_globe = pd.read_csv("https://data.giss.nasa.gov/gistemp/graphs/graph_data/Global_Mean_Estimates_based_on_Land_and_Ocean_Data/graph.txt",
                      skiprows=[0,1,2,4],
                      sep=r"\s+")

df_globe = df_globe.rename(columns={"No_Smoothing": "Mean Celsius Temperature (No Smoothing)",
                                   "Lowess(5)" : "Mean Celsius Temperature (Lowess Smoothing)"})

df_globe
```

```python
sns.set(style="ticks")

# Show the results of a linear regression within each dataset
ax = sns.lmplot(x="Year", y="Mean Celsius Temperature (No Smoothing)", data=df_globe, height=10)

plt.title("Source: NASA")

plt.suptitle("Global Mean Celsius Temperature Per Year, 1880 to 2020")

plt.show()
```
